```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

```python
df = pd.read_csv("otomoto.csv", sep=";")
df.head(5)
```

```python
df.dtypes
```

```python
df = df.drop(['Skrzynia biegów', 'Moc', 'Faktura VAT'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={"Marka pojazdu": "marka"})
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.count()   
```

```python
df = df[['marka', 'price', 'Rok produkcji']]
```

```python
df.head(10)
```

```python
df.count()
```

```python
df = df.dropna()
```

```python
df.count()
```

```python
sns.boxplot(x=df['price'])
```

```python
df.marka.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['Rok produkcji'], df['price'])
ax.set_xlabel('Rok produkcju')
ax.set_ylabel('cena')
plt.show()
```
